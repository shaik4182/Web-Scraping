```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
res = requests.get("https://www.flipkart.com/search?q=moblies&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
```

```python
res
```

```python
res.content
```

```python
soup = BeautifulSoup(res.content)
```

```python
soup
```

```python
products=[] #List to store name of the product
prices=[] #List to store price of the product
Rating =[]
specification = []
```

```python
for a in soup.findAll('a',href=True, attrs={'class':'_31qSD5'}):
    name=a.find('div', attrs={'class':'_3wU53n'})
    price=a.find('div', attrs={'class':'_1vC4OE _2rQ-NK'})
    rating=a.find('span',attrs={'class':'_38sUEc'})
    specific = a.find('div',attrs={'class':'_3ULzGw'})
    products.append(name.text)
    prices.append(price.text)
    Rating.append(rating.text)
    specification.append(specific.text)
```

```python
df = pd.DataFrame({'Product Name':products,'Price':prices,'Rating & Reviews':Rating,'Specification':specification}) 
df.to_csv('flipkart_moblie_data.csv', index=False, encoding='utf-8')
```
